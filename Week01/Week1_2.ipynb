You can then build if/then statements to set new variables.

An if/then statement has this structure.



```
if [insert condition]:
  indent next line, what you want to happen
elif [insert condition]:
  indent next line, what you want to happen
else:
  indent next line, what you want to happen
```

elif means "else if". You can have as many of these as you want, but you can only have one "if" and one "else".

Essentially, you are saying, if (**if**) this condition, then do A, or if (**elif**)this condition, do B, or if (**elif**) this condition, do C, etc, etc, until finally you say, if none of the above (**else**), do condition Z.



````python
# See this example
# We are checking if a is greater than b, equal, or something else.
a = 10
b = 12
if a > b:
  y = "a is greater than b"
elif a == b:
  y = "b is equal to a"
else:
  y = "b is greater than a"
print(y)
````

## 1.5 Loops

Loops are an essential part of programming. They allow you to cycle through a list of things of any length and apply some operation to them.

There are four essential parts of a loop:
1. *for* : this signifies that you are going to begin iterating through a loop
2. *iterator* : this is a temporary variable which will represent each iterative value
3. *range* : this is the things you want to iterate through
4. *operation* : this is what you will do in each iteration

Somewhat similar to if/then statements, the structure is:



```
for iterator in range:
  indent next line, complete operation
```




````python
for x in range(10): #x is the iterator, and range(10) is the range
  print(x) #print(x) is the operation. It says print the value of x to the console
````

````python
fruits = ["banana", "orange", "strawberry", "dragonfruit", "pineapple"]
for fruit in fruits: #fruit is the iterator, and fruits is the range
  print(fruit) #print(fruit) is the operation. It says print each fruit in the list to the console

````

````python
# You can combine for loops with other skills! For example, if/else conditions
# Here we will only print out values greater than 5
for x in range(10): # x is the iterator, and 1:10 is the range
  if x>5: #we are testing if x is greater than 5
    print(x) #this is the operation, we print it to the console
  else: #if x is not greater than 5
    continue # we continue to the next value in the range
````

***Tip:*** Everytime you use a : , the next line will be indented!

## 1.5 Functions

Functions allow you to write a set of code that you want to reuse easily.


````python
# For example, imagine you wanted to add 5 to a bunch of variables
a = 6
b = 7
c = 8
a = a + 5
b = b + 5
c = c + 5
````

This would get annoying and cumbersome if you had too many to do.

Functions make this easier. The basics of a function are:
1. *input* - the variable(s) you want to use to make a change
2. *output* - the variable(s) that come out the other side
3. *steps* - the things you need to do to get the input into the output

The basic structure is:



```
def function_name(input1, input2):
  indent next line, add steps
  return(output)
```




````python
def addFive(input): # This takes in a variable which we call input, you can name this anything you want
  newvalue = input + 5 # Goes through the steps of adding 5 to the input
  newvalue = newvalue + 0 # You can add as many steps as you want
  return(newvalue) # Returns the output back to you

# Notice when you run it, nothing happens in the console and no new variables are created
# You need to now "call" the function for it to work!
````

````python
print(a)
print(addFive(a))
print(a) # Notice this doesn't change the variable "a".
a = addFive(a) # When you set "a" to the output of the function, it does change it!
print(a)
````

````python
# A function can have multiple inputs
def createAddress(house, street, city, state):
  sep = " "
  address = str(house) + sep +  str(street) + sep +  str(city) + sep + str(state)
  return(address)


myaddress = createAddress(219, "West Sibley", "Ithaca", "NY")
print(myaddress)
````

````python
# It is good practice to add a comment inside your function
# So you know what it does!

def createAddress(house, street, city, state):
  """
  This function takes in different parts of an address
  and concatenates them together with a space separator.
  """
  sep = " "
  address = str(house) + sep +  str(street) + sep +  str(city) + sep + str(state)
  return(address)


myaddress = createAddress(219, "West Sibley", "Ithaca", "NY")
print(myaddress)
````

````python
# You can also have multiple outputs
def createAddress(house, street, city, state):
  sep = " "
  address = str(house) + sep +  str(street) + sep +  str(city) + sep + str(state)
  citystate = str(city) + sep + str(state)
  return address, citystate #separate the outputs with a comma

myaddress, mycity = createAddress(219, "West Sibley", "Ithaca", "NY")
print(myaddress)
print(mycity)
````
